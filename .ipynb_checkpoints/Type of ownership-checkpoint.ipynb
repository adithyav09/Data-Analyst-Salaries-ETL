```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from matplotlib import pyplot as plt
from scipy.stats import linregress
```

```python
csv_file = "cleaned_DA_salaries.csv"
data_df = pd.read_csv(csv_file)
data_df.head()
```

```python
#Clean up the-1 values and adding them to the "Unknown" group
data_df["Size"] = data_df["Size"].replace({'-1':'Unknown'})
company_size = data_df.groupby("Size").nunique()
company_size
```

```python
data_df = data_df.replace({'-1':np.NaN})
data_df.head(10)
```

```python
data_df = data_df.dropna()
data_df
```

```python
#Cleaning 'LoanRange' column by getting rid of unnecessary text
minimum_salary = data_df['Salary Estimate'].apply(lambda x: x.split('-')[0])
maximum_salary = data_df['Salary Estimate'].apply(lambda x: x.split('-')[1])
subtract_glassdoor_unit = maximum_salary.apply(lambda x: x.replace('(Glassdoor est.)', ''))
maximum_salary= subtract_glassdoor_unit
# #Separting loan range into minimum and maximum values 
# #Cleaning up commas and converting data type to integer
minimum_salary = minimum_salary.apply(lambda x: x.replace('K', ''))
maximum_salary = maximum_salary.apply(lambda x: x.replace('K', ''))
minimum_salary = minimum_salary.apply(lambda x: int(x.replace('$', '')))
maximum_salary = maximum_salary.apply(lambda x: int(x.replace('$', '')))
```

```python
Type_of_ownership = data_df[['Job Title', 'Salary Estimate', 'Size', 'Type of ownership']].copy()
Type_of_ownership.head()
```

```python
Type_of_ownership = data_df.groupby("Type of ownership").nunique()
Type_of_ownership.head(10)                                                
```

```python
data_df['Min Salary'] = minimum_salary
data_df['Max Salary'] = maximum_salary
data_df
```

```python
Type_of_ownership = data_df.groupby("Type of ownership").nunique()
Type_of_ownership                                                
```

```python
average_min_pay = data_df.groupby("Type of ownership").mean()['Min Salary']
average_max_pay = data_df.groupby("Type of ownership").mean()['Max Salary']
average_max_pay
```

```python
x = Type_of_ownership.index
plt.bar(x, average_min_pay)
plt.title("Average Minimum Salary for Type of ownership")
plt.xticks(x, rotation="vertical")
plt.ylabel("Minimum thousands of dollars")
plt.show()
```

```python
plt.bar(x, average_max_pay)
plt.title("Average Maximum Salary for Type of ownership")
plt.xticks(x, rotation="vertical")
plt.ylabel("Maximum thousands of dollars")
plt.show()
```
